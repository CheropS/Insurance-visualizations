```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('INSURGENCE FORM RESPONSES - Sheet1.csv')
```

```python
df = df.drop(columns=["First Name", "Email", "Phone Number", "Date Submitted", "Time Submitted"])
```

```python
duplicate_count = df.duplicated().sum()
print("Number of duplicate rows:", duplicate_count)

# Optionally drop duplicates
df = df.drop_duplicates()
```

```python
# --- Pie chart: Gender distribution ---
gender_counts = df["Gender"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(
    gender_counts, 
    labels=gender_counts.index, 
    autopct='%1.1f%%', 
    startangle=140
)
plt.title("Gender Distribution")
plt.show()
```

```python
# --- Pie chart: Gender distribution ---
disability_counts = df["Has Disability"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(
    disability_counts, 
    labels=disability_counts.index, 
    autopct='%1.1f%%', 
    startangle=140
)
plt.title("Has disability Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(y="Occupation", data=df, order=df["Occupation"].value_counts().index, palette="Set2")
plt.title("Distribution of Occupations in Dataset")
plt.xlabel("Count")
plt.ylabel("Occupation")
plt.show()
```

```python
#checking whether different occupations influence if they have purchased insurance 

# Cross-tabulation of Occupation vs. Has Purchased Insurance
ct = pd.crosstab(df['Occupation'], df['Has Purchased Insurance'])
print(ct)
```

```python
ct.plot(kind='bar', stacked=True, figsize=(12,6), colormap='Set2')
plt.title('Insurance Adoption Patterns by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Number of Respondents')
plt.legend(title='Has Purchased Insurance')
plt.xticks(rotation=45)
plt.show()
```

```python
# Split responses into lists
df['Reasons for Avoiding Insurance'] = df['Reasons for Avoiding Insurance'].str.split(',')

# Expand lists into separate rows
df_exploded = df.explode('Reasons for Avoiding Insurance')

# Clean whitespace
df_exploded['Reasons for Avoiding Insurance'] = df_exploded['Reasons for Avoiding Insurance'].str.strip()

ct = pd.crosstab(df_exploded['Reasons for Avoiding Insurance'], df_exploded['Has Purchased Insurance'])

ct.plot(kind='bar', stacked=True, figsize=(12,6), colormap='Set3')
plt.title('Insurance Adoption vs. Reasons for Avoiding Insurance')
plt.xlabel('Reasons for Avoiding Insurance')
plt.ylabel('Number of Respondents')
plt.legend(title='Has Purchased Insurance')
plt.xticks(rotation=45, ha='right')
plt.show()

```

```python

# Select company rating columns
company_cols = ["Company 1 Rating", "Company 2 Rating", 
                "Company 3 Rating", "Company 4 Rating", "Company 5 Rating"]

# Calculate average rating per company
avg_ratings = df[company_cols].mean()

# Plot line chart
plt.figure(figsize=(8,6))
sns.lineplot(x=avg_ratings.index, y=avg_ratings.values, marker='o')
plt.title("Average Company Ratings Comparison")
plt.ylabel("Average Rating")
plt.xlabel("Company")
plt.xticks(rotation=45)
plt.show()
```

# Advanced Visualizations

The following visualizations provide deeper insights into the insurance survey data.

```python
# 1. Bar chart: Age Group vs. Trust Rating
plt.figure(figsize=(12, 6))
age_trust = df.groupby('Age Group')['Trust Rating (1-5)'].mean().sort_values(ascending=False)
sns.barplot(x=age_trust.index, y=age_trust.values, palette='viridis')
plt.xlabel('Age Group', fontsize=12, fontweight='bold')
plt.ylabel('Average Trust Rating (1-5)', fontsize=12, fontweight='bold')
plt.title('Average Trust Rating by Age Group', fontsize=14, fontweight='bold', pad=20)
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 5)
# Add value labels on bars
for i, val in enumerate(age_trust.values):
    plt.text(i, val + 0.1, f'{val:.2f}', ha='center', va='bottom', fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# 2. Cross-tab bar chart: Age Group vs Has Purchased Insurance
plt.figure(figsize=(12, 6))
crosstab = pd.crosstab(df['Age Group'], df['Has Purchased Insurance'], normalize='index') * 100
# Melt for seaborn
crosstab_melted = crosstab.reset_index().melt(id_vars='Age Group', var_name='Has Purchased Insurance', value_name='Percentage')
sns.barplot(data=crosstab_melted, x='Age Group', y='Percentage', hue='Has Purchased Insurance', palette={'yes': '#2ecc71', 'no': '#e74c3c', 'not-sure': '#f1c40f'})
plt.xlabel('Age Group', fontsize=12, fontweight='bold')
plt.ylabel('Percentage (%)', fontsize=12, fontweight='bold')
plt.title('Insurance Purchase Status by Age Group', fontsize=14, fontweight='bold', pad=20)
plt.legend(title='Has Purchased Insurance', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 3. Stacked bar chart: Reasons for Avoiding Insurance vs Adoption
plt.figure(figsize=(14, 8))
data_list = []
for idx, row in df.iterrows():
    reasons = row['Reasons for Avoiding Insurance']
    # Fix: Strictly check for string to avoid ambiguity with array-like objects
    if isinstance(reasons, str):
        for reason in reasons.split(','):
            data_list.append({
                'Reason': reason.strip(),
                'Has Purchased': row['Has Purchased Insurance']
            })

reasons_df = pd.DataFrame(data_list)

# Use histplot for stacked bar effect with seaborn
if not reasons_df.empty:
    sns.histplot(data=reasons_df, y='Reason', hue='Has Purchased', multiple='stack', shrink=0.8, palette={'yes': '#2ecc71', 'no': '#e74c3c', 'not-sure': '#f1c40f'})
    plt.xlabel('Number of Responses', fontsize=12, fontweight='bold')
    plt.ylabel('Reasons for Avoiding Insurance', fontsize=12, fontweight='bold')
    plt.title('Reasons for Avoiding Insurance by Purchase Status', fontsize=14, fontweight='bold', pad=20)
    plt.legend(title='Has Purchased Insurance', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
```

```python
# 3. Stacked bar chart: Reasons for Avoiding Insurance vs Adoption
# ============================================
print("Creating Chart 3: Reasons for Avoiding Insurance...")
plt.figure(figsize=(14, 8))
# Split the reasons (they might be comma-separated)
data_list = []
for idx, row in df.iterrows():
    reasons = row['Reasons for Avoiding Insurance']
    if isinstance(reasons, str):
        for reason in reasons.split(','):
            data_list.append({
                'Reason': reason.strip(),
                'Has Purchased': row['Has Purchased Insurance']
            })

reasons_df = pd.DataFrame(data_list)

# Use histplot for stacked bar effect with seaborn
if not reasons_df.empty:
    sns.histplot(data=reasons_df, y='Reason', hue='Has Purchased', multiple='stack', shrink=0.8, palette={'yes': '#2ecc71', 'no': '#e74c3c', 'not-sure': '#f1c40f'})
    plt.xlabel('Number of Responses', fontsize=12, fontweight='bold')
    plt.ylabel('Reasons for Avoiding Insurance', fontsize=12, fontweight='bold')
    plt.title('Reasons for Avoiding Insurance by Purchase Status', fontsize=14, fontweight='bold', pad=20)
    plt.legend(title='Has Purchased Insurance', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
```

```python
# 4. Bar chart: Influential Platforms by Age Group
plt.figure(figsize=(14, 8))
platform_age_data = []
for idx, row in df.iterrows():
    if pd.notna(row['Influential Platforms']) and pd.notna(row['Age Group']):
        platforms = [p.strip() for p in str(row['Influential Platforms']).split(',')]
        for platform in platforms:
            platform_age_data.append({'Platform': platform, 'Age Group': row['Age Group']})

platform_df = pd.DataFrame(platform_age_data)
if len(platform_df) > 0:
    sns.countplot(data=platform_df, x='Platform', hue='Age Group', palette='tab10')
    plt.xlabel('Platform', fontsize=12, fontweight='bold')
    plt.ylabel('Number of Mentions', fontsize=12, fontweight='bold')
    plt.title('Influential Platforms by Age Group', fontsize=14, fontweight='bold', pad=20)
    plt.legend(title='Age Group', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
# 5. Radar chart: Reward Preferences

reward_cols = ['Reward: Reduced Premiums', 'Reward: Extra Loyalty', 
               'Reward: Recognition', 'Reward: VIP Access']
reward_scores = []
reward_labels = ['Reduced\nPremiums', 'Extra\nLoyalty', 'Recognition', 'VIP\nAccess']

for col in reward_cols:
    if col in df.columns:
        yes_count = df[col].value_counts().get('yes', 0)
        total = df[col].notna().sum()
        percentage = (yes_count / total * 100) if total > 0 else 0
        reward_scores.append(percentage)
    else:
        reward_scores.append(0)

fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(projection='polar'))
angles = np.linspace(0, 2 * np.pi, len(reward_labels), endpoint=False).tolist()
reward_scores_plot = reward_scores + [reward_scores[0]]
angles += angles[:1]

ax.plot(angles, reward_scores_plot, 'o-', linewidth=2, color='steelblue')
ax.fill(angles, reward_scores_plot, alpha=0.25, color='steelblue')
ax.set_xticks(angles[:-1])
ax.set_xticklabels(reward_labels, size=11, fontweight='bold')
ax.set_ylim(0, 100)
ax.set_yticks([20, 40, 60, 80, 100])
ax.set_yticklabels(['20%', '40%', '60%', '80%', '100%'])
ax.grid(True)
plt.title('Reward Preferences Distribution', fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()
```

```python
# 6. Bar chart: Preferred Lifestyle Reward Areas
plt.figure(figsize=(12, 6))
lifestyle_data = df['Preferred Lifestyle Reward Areas'].dropna()
all_areas = []
for areas in lifestyle_data:
    if isinstance(areas, str):
        all_areas.extend([a.strip() for a in areas.split(',')])

area_counts = pd.Series(all_areas).value_counts()
sns.barplot(x=area_counts.index, y=area_counts.values, palette='husl')
plt.xlabel('Lifestyle Reward Area', fontsize=12, fontweight='bold')
plt.ylabel('Number of Preferences', fontsize=12, fontweight='bold')
plt.title('Preferred Lifestyle Reward Areas', fontsize=14, fontweight='bold', pad=20)
plt.xticks(rotation=45, ha='right')

for i, val in enumerate(area_counts.values):
    plt.text(i, val + 0.5, str(val), ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()
```
